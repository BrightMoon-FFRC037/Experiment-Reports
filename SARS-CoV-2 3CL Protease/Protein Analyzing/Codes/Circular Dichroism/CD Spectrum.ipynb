```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Spectrum.xlsx')
data
```

```python
x=data['wave_blank']
y=data['ellipticity_blank']
plt.plot(x,y)
```

```python
lst = ['0','1.2','2.4','3.6','4.8','6']
for i in lst:
    x=data['wave_'+f'{i}']
    y=data['ellipticity_'+f'{i}']
    plt.plot(x,y,label=f'c(guanidine)={i}M')
plt.legend()
```
