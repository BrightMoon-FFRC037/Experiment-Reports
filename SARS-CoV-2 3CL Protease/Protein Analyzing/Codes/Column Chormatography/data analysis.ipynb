```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('./Affinity Column-B Concentration.xlsx')
volume = np.array(data['ml'])
y = np.array(data['c(B)'])
plt.plot(volume,y,label='B Percentage')
plt.xlabel('Volume (mL)')
plt.ylabel('Percentage of Buffer B (%)')
plt.legend()
ax_1 = plt.gca().twinx()
data_1 = pd.read_excel('./Affinity Column-UV and Conductance.xlsx')
volume_1 = np.array(data_1['ml_UV'])
y = np.array(data_1['mAU'])
ax_1.plot(volume_1,y,color='orange',label='UV Absorbance')
plt.ylabel('280 nm UV Absorbance (mAU)')
plt.legend()
plt.ylim(-200,3700)
```
